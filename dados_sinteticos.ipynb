```python
import datetime
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.reload_library()
plt.style.use(['science','ieee','no-latex'])
plt.rcParams.update({
    "legend.fancybox": False,
    "legend.frameon": False,
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Times"],
    "font.size":11})
```

```python
DATA_DIR = os.path.join('data', 'sinteticos')
```

```python
dates = pd.date_range(start=datetime.date(2020, 12, 14) - datetime.timedelta(days=250), end=datetime.date(2020, 12, 13))
```

```python
trend = np.arange(250) * 10
```

```python
time = np.arange(50)
values = np.where(time < 10, time**3, (time - 9)**2)

seasonality = []
for i in range(5):
    for j in range(50):
        seasonality.append(values[j])
```

```python
noise = np.random.randn(250) * 100
```

```python
big_event = np.zeros(250)
big_event[-50:] = np.arange(50) * -50
```

```python
synthetic = trend + seasonality + noise + big_event
```

```python
df = pd.DataFrame(data=synthetic, columns=['valor'], index=dates)
trend_df = pd.DataFrame(data=trend, columns=['valor'], index=dates)
seasonality_df = pd.DataFrame(data=seasonality, columns=['valor'], index=dates)
noise_df = pd.DataFrame(data=noise, columns=['valor'], index=dates)
big_event_df = pd.DataFrame(data=big_event, columns=['valor'], index=dates)
```

```python
trend_df.plot(figsize=(10, 5));
seasonality_df.plot(figsize=(10, 5));
noise_df.plot(figsize=(10, 5));
big_event_df.plot(figsize=(10, 5));
```

```python
df.to_csv(os.path.join(DATA_DIR, 'sinteticos.csv'), index=False, sep=';', date_format='%Y-%m-%d')
```
