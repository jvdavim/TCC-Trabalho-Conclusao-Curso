# Exploratory data analysis - Minimum daily temperatura in Melbourne

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.reload_library()
plt.style.use(['science','ieee','no-latex'])
plt.rcParams.update({
    "legend.fancybox": False,
    "legend.frameon": False,
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Times"],
    "font.size":11})
```

```python
ts = pd.read_csv('../data/sinteticos/sinteticos.csv', index_col=0, sep=';')

fig, ax = plt.subplots()
ts.loc[:, 'valor'].plot(figsize=(10, 5), ax=ax)
ax.legend(["Valor"]);
```

```python
ts['moving_average'] = ts.rolling(4).mean().valor

fig, ax = plt.subplots()
ts.plot(figsize=(10, 5), ax=ax)
ax.legend(["Número de casos confirmados", "Médias móveis"]);
```
